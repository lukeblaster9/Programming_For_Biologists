```python
from collections import defaultdict
import csv

#declare some dictionaries
presidents = defaultdict(str) # key =president number, value = president name
party_count = defaultdict(int) # key = party, value = ncount of presidents in that party
#without defaultdict, we'd declare a dictionary this way: parties = {}
#party_count = {}

#open the presidents.csv data file
input_file = "../Data/presidents.csv"

#open the file
with open(input_file, "r") as prez_file:
   #create a csv reader object
    reader = csv.reader(prez_file, delimiter= ",")
   
   #skip the header line
    header= next(reader)

    #read the file line by line
    for line in reader:
        #print just to make sure we're reading the file okay
        #print(line)

        # strip the line breaks = depracated
        #line = line.strip()

        #split each line into a list = depracated and replacing with csv reader
        #columns = line.split(',')

        #did we split it?
        print(line[0])

        #load our presidents{} dictionary
        num = line[0].strip()
        name = line[1].strip()
        presidents[num] =name
        # --- OR --- #
        #presidents[columns[0]] = columns[1]

        #load the party_count dictionary
        party = line[5].strip()
        party_count[party] += 1
```

```python
#loop over our party county dictionary
for party, count in party_count.items():
    print(f"{party}: {count}")
```
