### Populate and print a list

```python
beatles = ["John", "Paul", "George", "Ringo", "Stuart"]
#print the whole list
print(beatles)
print(beatles[0:3])
```

### Calculate and print the length of a list

```python
# calculate number of beatles
number_of_beatles=len(beatles)
# print number of beatles
print(f"There were {number_of_beatles} members of the Beatles.")
```

### Print an item in a list 
   
   _see PowerPoint slides_

```python
# print the first beatle
print(beatles[0])
# print the last Beatle
print(beatles[4])

#Many ways to print the last Beatle
print(beatles[-1])
print(beatles[len(beatles)-1])

#Print George
print(beatles[-3])
print(beatles[2])
print(beatles[len(beatles)-3])
```

### Remove an item from a list

```python
# Remove the last item in the list (bye, Stuart)
stu = beatles.pop()
# What happened to the Beatles?
print(stu)
print(beatles)
```

### Make a new list

```python
# make a list called 'stones' with the Rolling Stones (Mick, Ron, and Keith)
stones = ["Mick", "Keith", "Ron"]
```

### Add two lists

```python
#Print the two groups
print("The two groups separately: {}, {}".format(beatles, stones))
print()
# add two lists together
supergroup = beatles + stones
# print our new list
print(f"{len(supergroup)} members in our new band: {supergroup}")
```

### Growing a list with `extend()`

```python
# print our lists
print("Jackson 5:", jackson5)
print("Destiny's Child:", destiny)
# use extend() to add Destiny's Child to the Jackson 5
jackson5.extend(destiny)
#Here's what it's doing:
#jackson5 = jackson5 + destiny

print(f"Here's the extended Jackson5: {jackson5}")
```

### Growing a list with `append()`

```python
# use `append()` to add lists in a different way
num = [1, 2, 3]
let = ["a", "b", "c"]
print(f"our lists: {let, num}")

#append let onto num
num.append(let)

# What did append do?
print(f"after appending let to num: {num}")

more_num = [4, 5, 6]
more_num.append(7)
i = 8
more_num.append(i)
print(more_num)
```

```python
# How long is this list?
print(len(num))

#print the letters
print(num[-1])
print(num[3])

#find and print 'b'
#Here's how to get the list
print(num[3])

#Here's how to get an item i nthe list, 'b' for example
print(num[3][1])
```

### Strings as list

- strings are just a list of characters
- this means you can index into a tring, the same way you would a list



```python
name = 'beyonce'
print(name[0], name[-1])

#Here's a gene with structure EXONintronEXON
gene = "ATGCCAGTttccggaaGTTCTAA"

#Extract the first exon (i.e., the first string of capital letters: "ATGCCAGT")
print(f"the first exon: {gene[0:8]}")

#Get the start codon (the first three characters of the gene)
print(gene[0:3])

#Get the stop codon (the last three characters of the gene)
print(gene[-3:])

#Get the last exon (i.e., the last string of capital letters: "GTTCTAA")
print(gene[-7:])
```
